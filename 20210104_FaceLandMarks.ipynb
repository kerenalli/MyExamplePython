```
#Python Code to detect the mouth portion.
import cv2
import numpy as np
import dlib
# Load the detector
detector = dlib.get_frontal_face_detector()
# Load the predictor
predictor = dlib.shape_predictor("/content/shape_predictor_68_face_landmarks.dat")
# read the image
img = cv2.imread("/content/face.png")
# Convert image into grayscale
gray = cv2.cvtColor(src=img, code=cv2.COLOR_BGR2GRAY)
# Use detector to find landmarks
faces = detector(gray)
for face in faces:
    x1 = face.left() # left point
    y1 = face.top() # top point
    x2 = face.right() # right point
    y2 = face.bottom() # bottom point
    # Create landmark object
    landmarks = predictor(image=gray, box=face)
    # Loop through all the points
    for n in range(48, 68):
        x = landmarks.part(n).x
        y = landmarks.part(n).y
        # Draw a circle
        cv2.circle(img=img, center=(x, y), radius=3, color=(0, 255, 0), thickness=-1)
# show the image
plt.imshow(img)
# Delay between every fram
cv2.waitKey(delay=0)
# Close all windows
cv2.destroyAllWindows()
```

```
#Resizing of an image in Python with OpenCV
h1=400
w1=400
dimension = (w1, h1) 
resized_image = cv2.resize(image, dimension, interpolation = cv2.INTER_AREA)
plt.imshow(resized_image)
```

```
#Cropping of an image in Python
cropped_image = image[210:240, 300:360]
h1=256
w1=256
dimension = (w1, h1) 
resized_image = cv2.resize(cropped_image, dimension, interpolation = cv2.INTER_AREA)
plt.imshow(resized_image)
plt.imshow(cropped_image)
plt.imsave("cimage.jpg",cropped_image)
img= plt.imread('/content/cimage.jpg')
plt.imshow(img)
```

```
#Rotation of an image in OpenCV
(h1, w1) = image.shape[:2]
center = (w1 / 2, h1 / 2)
Matrix = cv2.getRotationMatrix2D(center, 180, 1.0)
rotated_image = cv2.warpAffine(image, Matrix, (w1, h1))
plt.imshow(rotated_image)
```
